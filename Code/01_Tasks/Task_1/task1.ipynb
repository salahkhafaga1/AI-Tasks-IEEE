```python
def matsum(arr1, arr2):
    result = []
    for i in range(len(arr1)):
        row = []
        for j in range(len(arr1[0])):
            row.append(arr1[i][j] + arr2[i][j])
        result.append(row)
    return result

```

```python
def matsub(arr1, arr2):
    result = []
    for i in range(len(arr1)):
        row = []
        for j in range(len(arr1[0])):
            row.append(arr1[i][j] - arr2[i][j])
        result.append(row)
    return result

```

```python
def matmul(arr1, arr2):
    result = []
    for i in range(len(arr1)):
        row = []
        for j in range(len(arr2[0])):
            sum_product = 0
            for k in range(len(arr2)):
                sum_product += arr1[i][k] * arr2[k][j]
            row.append(sum_product)
        result.append(row)
    return result


```

```python
def scalarsum(scalar, arr):
    result = []
    for i in range(len(arr)):
        row = []
        for j in range(len(arr[0])):
            row.append(arr[i][j] + scalar)
        result.append(row)
    return result

```

```python
def scalarsub(scalar, arr):
    result = []
    for i in range(len(arr)):
        row = []
        for j in range(len(arr[0])):
            row.append(arr[i][j] - scalar)
        result.append(row)
    return result


```

```python
def scalarmul(scalar, arr):
    result = []
    for i in range(len(arr)):
        row = []
        for j in range(len(arr[0])):
            row.append(arr[i][j] * scalar)
        result.append(row)
    return result


```

```python
def matnorm(arr):
    flat_list = []
    for row in arr:
        for val in row:
            flat_list.append(val)
            
    min_val = min(flat_list)
    max_val = max(flat_list)
    diff = max_val - min_val

    result = []
    for i in range(len(arr)):
        row = []
        for j in range(len(arr[0])):
            if diff == 0:
                row.append(0.0)
            else:
                norm_value = (arr[i][j] - min_val) / diff
                
                row.append(round(norm_value, 2))
        result.append(row)
    return result

```
