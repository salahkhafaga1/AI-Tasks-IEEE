# Part 1

**You are given the following dataset representing students’ grades in 4 subjects:**

```python
import numpy as np
```

```python
grades =[[85, 78, 92, 88],
[70, 76, 80, 65],
[90, 88, 94, 91],
[60, 65, 58, 62],
[100, 95, 98, 97]] 
```

**Convert the data into a NumPy array.**


```python
grades_array = np.array(grades)
```

**shape of array**

```python
print(f"the shape of the array is: {grades_array.shape}")
```

**The mean grade of each student**
**The mean grade of each subject.**

```python
# *The mean grade of each student*
the_mean = np.mean(grades_array)
print(f"the mean of the grades is: {the_mean}")
# **The mean grade of each subject.**
the_mean_subject = np.mean(grades_array, axis=0)
print(f"the mean of the grades for each subject is: {the_mean_subject}")
```

**• Extract the students whose average grade is greater than 85.**

```python
the_avr_subject = np.average(grades_array, axis=1)
for i in range(len(the_avr_subject)):
    if the_avr_subject[i] >85:
        print(f"the average grade of student {i+1} is: {the_avr_subject[i]} and it is above 85")
    
```

**Add a bonus of 5 marks to all grades using broadcasting.**

```python
# Add a bonus of 5 marks to all grades using broadcasting. 
grades_array += 5
print(f"the grades after adding the bonus are: \n{grades_array}")
```

**• Normalize the grades using Min-Max normalization.**

```python
 
min_grades = np.min(grades_array)
max_grades = np.max(grades_array)
normalized_grades = (grades_array - min_grades) / (max_grades - min_grades)
print(f"the normalized grades are: \n{normalized_grades}")
```

**Flatten the array into a single vector.**

```python
 
flattened_grades = grades_array.flatten()
print(f"the flattened grades are: \n{flattened_grades}")
```

# Part 2

**Linear Regression from Scratch (Normal Equation)**

```python
house_size=[50, 60, 80, 100, 120]  
price_in_thousands=[150, 180, 240, 300, 330]  
```

```python
# convert the lists to numpy arrays
x=np.array(house_size)
y=np.array(price_in_thousands)
```

```python
# reshape the arrays to be 2D
x=x.reshape(-1, 1)
y=y.reshape(-1, 1)
```

```python
# • Add a bias column of ones to X.
X = np.hstack((x, np.ones((x.shape[0], 1))))
X
```

```python
print(f"the shape of X is: {X.shape} \n the shape of y is: {y.shape}")
```

```python
# Compute the parameter vector θ using the Normal Equation: θ = (X^T X)^(-1) X^T y.
theta = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
# • Use the computed θ to predict the price of a 90 m² house. 
new_house = np.array([[90, 1]])  # 90 m² with bias term
predicted_price = new_house @ theta
print(f"the predicted price of a 90 m² house is: {predicted_price[0][0]} thousand dollars")
```
